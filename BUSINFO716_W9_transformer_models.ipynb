```python
from transformers import pipeline
```

```python
# classify a statement from the Air New Zealand 2022 sustainability report and a placebo

#sentence = "Air New Zealand’s employee engagement score being in Glint's Global Top 20% Engagement Index."
#sentence = "There can be few people who now doubt that accelerating climate change represents a massive challenge for governments and businesses the world over. But not many yet understand the true nature of this crisis, with disruption from more and more climate-induced disasters becoming ever more deadly – and costly." 
sentence = "Establish a baseline of Air New Zealand spend with Māori and Pasifika-owned businesses and social enterprises by 2022."
#sentence = 'this is an annual sustainability report'
```

```python
# tested in transformers==4.18.0 
from transformers import BertTokenizer, BertForSequenceClassification, pipeline

finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-esg',num_labels=4)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-esg')
nlp = pipeline("text-classification", model=finbert, tokenizer=tokenizer)
results = nlp(sentence)
print(results) 
```

```python
sentiment = pipeline("sentiment-analysis", model="facebook/bart-large-mnli")
sentiment(sentence)
```
